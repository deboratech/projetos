```python
ufo = pd.read_csv('/content/ufo.csv', encoding='utf-8')
ufo = ufo.drop_duplicates()
display(ufo)
```

```python
ufo.info()
```

```python
ufo.rename (columns={'posted':'postado'}, inplace = True)
ufo.rename (columns={'date':'data'}, inplace = True)
ufo.rename (columns={'time':'hora'}, inplace = True)
ufo.rename (columns={'city':'cidade'}, inplace = True)
ufo.rename (columns={'state':'estado'}, inplace = True)
ufo.rename (columns={'shape':'forma'}, inplace = True)
ufo.rename (columns={'duration':'duracao'}, inplace = True)
ufo.rename (columns={'summary':'resumo'}, inplace = True)
ufo.rename (columns={'images':'imagens'}, inplace = True)
ufo.rename (columns={'img_link':'link da imagem'}, inplace = True)
ufo.rename (columns={'lat':'lat'}, inplace = True)
ufo.rename (columns={'lng':'lng'}, inplace = True)
ufo.rename (columns={'population':'populacao'}, inplace = True)
```

```python
ufo.info()
```

```python
ufo = ufo.replace([float('NaN')],'-')
```

```python
ufo.dropna()
```

```python
ufo.info()
```

```python
display(ufo)
```

```python
formas_especificas = ['OVAL', 'TRIANGLE', 'STAR', 'CIGAR', 'DISK', 'CIRCLE', 'DIAMOND', 'EGG', 'CROSS', 'CYLINDER', 'RECTANGLE', 'CONE']
ufo_formas_especificas = ufo[ufo['forma'].str.upper().isin(formas_especificas)]
contagem_formas = ufo_formas_especificas['forma'].value_counts()
```

```python
plt.figure(figsize=(12, 6))
contagem_formas.plot(kind='barh', color='skyblue')
plt.title('Formas de Avistamentos de OVNIs')
plt.xlabel('Nmero de Avistamentos')
plt.ylabel('Forma')
plt.show()
```

```python
top_10_estados = ['IL','OR','WA','NY','AZ','PA','TX','CA', 'FL', 'OH']
ufo_top_estados = ufo[ufo['estado'].isin(top_10_estados)]
contagem_top_estados = ufo_top_estados['estado'].value_counts()
```

```python
plt.figure(figsize=(12, 6))
contagem_top_estados.plot(kind='barh', color='green')
plt.title('Nmero de Avistamentos de OVNIs nos 10 Estados Mais Relevantes')
plt.xlabel('Nmero de Avistamentos')
plt.ylabel('Estado')
plt.show()
```

```python
ufo = ufo[ufo['duracao'] != '------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------']
ufo['duracao'] = pd.to_numeric(ufo['duracao'], errors='coerce')
media_duracao = ufo['duracao'].mean()
print(media_duracao)
```

```python
plt.hist(ufo['duracao'].dropna(), bins=30, color='skyblue', alpha=0.7, label='Durao')
plt.axvline(x=media_duracao, color='red', linestyle='--', linewidth=2, label=f'Mdia: {media_duracao:.2f} segundos')
plt.title('Durao de Avistamentos de OVNIs')
plt.xlabel('Durao (segundos)')
plt.ylabel('Frequncia')
plt.legend()
plt.show()
```

```python
cidades_desejadas = ['Tucson', 'Phoenix', 'Chicago', 'Las Vegas']
ufo_cidades_desejadas = ufo[ufo['cidade'].isin(cidades_desejadas)]
contagem_cidades = ufo_cidades_desejadas['cidade'].value_counts()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(contagem_cidades, labels=contagem_cidades.index, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4), colors=plt.cm.Paired.colors)
plt.title('Avistamentos de OVNIs em Cidades Especficas')
plt.show()
```

```python
ufo['hora'] = pd.to_datetime(ufo['hora'], errors='coerce')
ufo['hora_do_dia'] = ufo['hora'].dt.hour
contagem_horas = ufo['hora_do_dia'].value_counts().sort_index()
media_duracao_por_hora = ufo.groupby('hora_do_dia')['duracao'].mean()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(contagem_horas.index, contagem_horas.values, 'r.--', label='Nmero de Avistamentos')
plt.axhline(y=contagem_horas.mean(), color='b', linestyle='--', label=f'Mdia: {contagem_horas.mean():.2f}')
plt.title('Avistamentos de OVNIs por Hora do Dia')
plt.xlabel('Hora do Dia')
plt.ylabel('Nmero de Avistamentos')
plt.legend()
plt.show()
```
